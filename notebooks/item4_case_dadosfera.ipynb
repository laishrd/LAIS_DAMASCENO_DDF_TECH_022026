```python

!pip uninstall -y great-expectations
!pip install great-expectations==0.15.50

import pandas as pd
import great_expectations as ge


df = pd.read_csv("products_and_descriptions.csv")

ge_df = ge.from_pandas(df)


```

```python
#nulo
ge_df.expect_column_values_to_not_be_null("docid")
ge_df.expect_column_values_to_not_be_null("title")
ge_df.expect_column_values_to_not_be_null("text")
```

```python
#unico
ge_df.expect_column_values_to_be_unique("docid")
ge_df.expect_column_values_to_be_unique("title")
ge_df.expect_column_values_to_be_unique("text")
```

```python
#tamanho minimo esperado = 20
ge_df.expect_column_value_lengths_to_be_between(
    "text",
    min_value=20
)
```

```python
#descrições invalidas ou incompletas
ge_df.expect_column_values_to_not_be_in_set(
    "text",
    ["Product Description", "N/A", ""]
)
```

```python
#relatorio
results = ge_df.validate()
results
```
